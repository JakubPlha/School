## 1. Jaká data jsou v description

- Jsou tam data z json. K ukládání infa o description.
- Je to ve stylu klíč: hodnota.

## 2. Co udělá dotaz
SELECT
name,
JSON_VALUE(description, '$.category') AS category
FROM product; 

name, JSON_VALUE - Vypíše název produktu a z JSON sloupce description vybere věci z category.
JSON_VALUE() - Slouží k získání konkrétní hodnoty z JSON. Výsledkem je teda seznam produktu a jejich kategorii.

## 3. SQL pro zobrazení všech produktů z vybraného skladu.

```python
SELECT *
FROM product
WHERE warehouse_id = 2;
```

## 4. SQL pro zobrazení všech produktů, které mají ve sloupci description v category uvedeno cable.

```python
SELECT *
FROM product
WHERE JSON_VALUE(description, '$.category') = 'cable';
```

## 5. SQL pro zobrazení všech produktů, jejichž počet je větší než 20 seřezený vzestupně podle počtu.

```python
SELECT *
FROM product
WHERE quantity > 20
ORDER BY quantity ASC;
```

## 6. SQL pro zobrazení všech produktů, které byly uskladněny v březnu 2024.

```python
SELECT * FROM product WHERE stored_from BETWEEN '2024-03-01' AND '2024-03-31';
```
